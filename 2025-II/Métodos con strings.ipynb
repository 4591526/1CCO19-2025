### **`str.lower()`**

Este método convierte todo a minúsculas.

```python
palabra = "APROBADO"
texto = "El 22 de febrero se nos anunció que regresaríamos a Colombia."

print(palabra)
print(texto)
```

```python
print(palabra.lower())
print(texto.lower())
```

```python
print(palabra)
palabra_minuscula = palabra.lower()
print(palabra_minuscula)
```

### **`str.upper()`**

Este método convierte todo a mayúsculas.

```python
palabra = "viernes"
```

```python
print(palabra.upper())
print(texto.upper())
```

### **`str.startswith()`**

Este método devuelve True si la cadena de caracteres empieza con el caracter o palabra indicada.

```python
print(palabra.startswith("v"))
print(texto.startswith("Ella"))
```

### **`str.endswith()`**

Este método devuelve True si la cadena de caracteres termina con el caracter o palabra indicada.

```python
print(palabra.endswith("A"))
print(texto.endswith("Colombia."))
```

### **`str.count()`**

Este método cuenta cuántas veces aparece un caracter o palabra en la cadena de caracteres principal.

```python
print(palabra.count("a"))
print(texto.count("re"))
```

### **`len(string)`**

Este función devuelve la cantidad de caracteres en el string, incluyendo espacios, signos de puntuación y saltos de línea.

```python
print(len(palabra))
print(len(texto))
```

---
### **`str.find()`**
Este método sirve para buscar un caracter o palabra dentro de la cadena de caracteres principal. Es decir, devuelve la posición donde empieza la cadena que estás buscando. Si no la encuentra, devuelve -1.

```python
texto = "Me gusta programar en Python"
print(texto.find("Python"))
print(texto.find("python"))
```

### **`str.index()`**
Este método sirve para buscar un caracter o palabra dentro de la cadena de caracteres principal y devuelve el índice donde empieza la primera coincidencia. Si no lo encuentra, devuelve un Error.

Sintaxis: **str.index(subcadena, inicio=0, fin=len(cadena))**

* subcadena: lo que quieres buscar

* inicio (opcional): desde qué índice comenzar a buscar

* fin (opcional): hasta qué índice buscar

```python
texto = "Me gusta programar en Python"
print(texto.index("Python"))
print(texto.index("Python", 20, 28))
print(texto.index("python"))
```

---
### **`str.replace(viejo, nuevo)`**

Este método reemplaza un caracter o una palabra en la cadena de caracteres principal.

```python
print(palabra.replace("e","a"))
print(texto.replace("Python", "Perú"))
```

### **`str.split(sep)`**

Este método divide el string en partes, usando un separador o no.

```python
print(palabra.split())
print(texto.split())
```
