### **`list.append()`**
Este método permite agregar un elemento al final de la lista.

```python
frutas = ["manzana", "pera", "uva"]

frutas.append("naranja")
print(frutas)
```

```python
frutas.append("manzana")
print(frutas)
```

### **`list.extend()`**
Este método permite extender la lista con otra lista.

```python
frutas.extend(["durazno", "sandia"])
frutas_2 = ["kiwi", "lúcuma"]
frutas_2.extend(frutas)
print(frutas)
print(frutas_2)
```

### **`list.insert(posición, elemento)`**

* posición → índice donde quieres insertar el nuevo elemento

* elemento → valor que se va a insertar en esa posición

Este método permite elegir en qué lugar exacto de la lista se coloca el nuevo elemento.


```python
frutas_2.insert(3, "papaya")
print(frutas_2)
```

### **`list.remove()`**
Este método elimina la primera ocurrencia de un elemento específico en la lista.


```python
frutas_2.remove("manzana")
print(frutas_2)
```

### **`list.pop(índice)`**
Este método elimina y devuelve un elemento de la lista.
Si no le indicas un índice, elimina el último elemento.





```python
último = frutas.pop()
print(último)
print(frutas)

segundo = frutas.pop(1)
print(segundo)
print(frutas)
```

### **`list.sort()`**
Este método ordena los elementos de una lista en orden ascendente por defecto.


```python
frutas_2.sort() # frutas_2.sort(key=None, reverse=False)
print(frutas_2)

frutas_2.sort(reverse=True) # frutas_2.sort(key=None, reverse=True)
print(frutas_2)

frutas_2.sort(key=len, reverse=False) # key → parámetro opcional que define cómo comparar los elementos
print(frutas_2)
```

### **`list.reverse()`**
Este método invierte el orden de los elementos de una lista, sin ordenarlos, solo los pone en el sentido contrario.

```python
print(frutas)

frutas.reverse()
print(frutas)
```

### **`list.count()`**

Este método cuenta cuántas veces aparece un elemento en una lista.

```python
numeros = [1, 2, 3, 2, 4, 2, 5]
print(numeros.count(2))
```

### **`list.index(elemento, inicio, fin)`**

Este método sirve para encontrar la posición (índice) de la primera aparición de un elemento en una lista.

* Si el elemento está varias veces, devuelve la primera posición.

* Si no existe, genera un ValueError.


    - elemento → el valor que buscas

    - inicio (opcional) → desde qué índice empezar a buscar

    - fin (opcional) → hasta qué índice buscar

```python
print(frutas_2)
```

```python
print(frutas_2.index("sandia"))

print(frutas_2.index("kiwi", 2, 6))

print(frutas_2.index("papaya"))
```

### **`list.clear()`**

Este método sirve para vaciar completamente una lista, dejándola vacía ([]).

* No elimina la variable, solo borra los elementos.



```python
frutas_2.clear()
print(frutas_2)
```
