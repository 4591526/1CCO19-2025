## **2. print()**
Esta función permite mostrar un texto especifico en la pantalla.
El texto a mostrar se escribe como **argumento** de la función.

```python
# Un argumento puede escribirse con comillas dobles
print("¡Hola mundo!")
```

```python
# Un argumento puede escribirse con comillas simples
print('¡Hola Mundo')
```

La función **`print()`** admite varios argumentos seguidos. En el programa, los argumentos deben separarse por comas.

```python
# La función print() puede contener más de un argumento
# La coma que separa cada argumento representa un espacio vacío
print("Hola", "Adiós")
```

```python
# Los espacios vacíos dentro de un argumento también se respetan cuando usamos la función print()
print("Hola Adiós") # el resultado es igual a print("Hola", "Adiós")
```

```python
# Sin una coma que separe argumentos se interpreta como un solo argumento sin espacio entre ellos
print("Hola" "Adiós")
```

```python
# Si se busca imprimir un argumento que contiene dos palabras sin espacio entre ellas se mostrará junto
print("HolaAdiós") # el resultado es igual a print("Hola" "Adiós")
```

Al final de cada **`print()`**, Python añade automáticamente un salto de línea.

```python
# Imprimamos dos argumentos distintos con print() en dos líneas de código
print("Aula")
print("Z104")
```

```python
# Agregamos un print() sin un argumento
print("Aula")
print()  # Python respeta el vacío como un espacio en blanco
print("Z104")
```

Si se quiere que Python no añada un salto de línea al final de un **`print()`**, se debe añadir al final el argumento **`end=""`**.

```python
# Se puede usar el párametro end vacío (end = "") para indicar que no deseamos imprimir en dos líneas distintas
print("Aula", end="")
print("Z104") # el resultado es igual a print("AulaZ104") o print("Aula" "Z104")
```

Si se quieren separar los argumentos en la salida, hay que incluir los espacios deseados.

```python
# Se puede agregar un espacio vacío al final del argumento de la primera línea
print("Aula ", end="")
print("Z104") # el resultado es igual a print("Aula", "Z104") o print("Aula Z104")
```

```python
# Se puede agregar un espacio vacío antes del argumento de la segunda línea
print("Aula", end="")
print(" Z104") # el resultado es igual a print("Aula", "Z104") o print("Aula Z104")
```

```python
print("Aula Z104")
```

También se puede ejecutar dos argumentos y especificar un separador.

```python
# Se puede indicar qué tipo de separador deseas entre los argumentos, estos se imprimiran en vez de indicar un espacio vacío por defecto
print("Hola", "¿Cómo están?", sep="--")
```

```python
print("Hola", "¿Cómo están?", sep=";")
```

```python
print("Hola", "¿Cómo están?", sep=":")
```

```python
print("Hola", "¿Cómo están?", sep=": ") # Puedes agregar un espacio vacío al final o al inicio de un argumento, también puedes agregarlo en el separador
```

## **3. help()**
Es una función integrada que proporciona información sobre funciones. Es útil para recuperar información sobre diversos objetos de Python.

```python
help(print)
```

### **`print(*args, sep=' ', end='\n')`**

* Admite argumentos
* sep: Se inserta una cadena entre los valores; por defecto, un espacio.
* end: Se añade una cadena después del último valor; por defecto, una nueva línea (\n).


```python
print("Mundial", 2026, sep=" \t", end='\n')
# \n  -> esto indica un salto de línea
# \t -> esto indica una sangría
```

## **4. Escribir comentarios en Python**
¿Viste que usamos el símbolo **`#`** para añadir un comentario al código? Los comentarios son líneas de texto que no se ejecutan como código, sino que se utilizan para hacer que el código sea más fácil de entender. Puedes añadir comentarios a tu código para recordar lo que hace, o para ayudar a otros a entenderlo.

Para escribir comentarios en Python, use el símbolo numeral **`#`** antes de escribir tu comentario. Cuando ejecutes tu código, Python ignorará todo lo que esté más allá del **`#`**.

```python
# Escribamos un comentario

print('Facultad de Ciencias y Artes de la Comunicación') # Esta línea imprime un string

print('Hola')
```

```python
# ¿Qué carreras se imprimen?

print("Publicidad")
# print("Periodismo")
print("Comunicación Audiovisual")
print("Comunicación para el Desarrollo")
```

## **5. Errores en Python**
Todos cometemos errores. Para muchos tipos de errores, Python les dirá que han cometido un error al darle un mensaje de error. Es importante leer los mensajes de error detenidamente para comprender realmente dónde se cometió un error y cómo puedes corregirlo.

Por ejemplo, si deletrea **print** como **frint**, Python mostrará un mensaje de error.

```python
# Ejecutamos una cadena de caracteres (string) como mensaje de error

frint("Hello, Python!")
```

```python
# Corregimos el código

print("Hello, Python!")
```

```python
# ¿Cuál es el error aquí?

print("Hello, Python!)
```

```python
# Corregimos el código

print("Hello, Python!")
```

Como las comillas indican el principio y el final de una cadena, si se escriben comillas dentro de comillas se produce un error de sintaxis.

```python
print("Una persona le dice a otro: "Hola"")
```

```python
# Corregimos el código anterior
print("Una persona le dice a otro: 'Hola'") # En este caso, solo tenemos un argumento y se diferencia las comillas para que Python no lo vea como error de sintaxis
```

```python
print("Una persona le dice a otro:", 'Hola') # En este caso, tenemos dos argumentos y se pueden usar en ambos casos las mismas comillas
```


### **¿Python sabe acerca de tu error antes de ejecutar tu código?**
Python es lo que se llama un lenguaje interpretado. Por su parte, los lenguajes compilados (C++, Java, entre otros) examinan todo el programa en el momento de la compilación y pueden advertirle sobre toda clase de errores antes de la ejecución. Por el contrario, Python interpreta línea por línea a medida que lo ejecuta. Python dejará de ejecutar todo el programa cuando encuentre un error.

```python
# Recorremos los siguientes strings para ver el orden de ejecución

print("Esto se ejecutará")
frint("Esto tendrá un error") # Python deja de ejecutar el siguiente print() porque encuentra un error en esta línea del tipo error en el nombre de la función
print("Esto NO se ejecutará")
```
