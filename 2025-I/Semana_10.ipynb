## **1. Base de Datos**
Una base de datos es un conjunto organizado de datos que permiten almacenar grandes cantidades de datos de manera estructurada. Esto facilita el acceso y la manipulación de la información.

Almacenar datos en una base de datos permite acceder a ellos de manera rápida y eficiente. Esto es especialmente útil para análisis que requieren consultar múltiples fuentes de información.

Las bases de datos permiten realizar consultas complejas para identificar tendencias y patrones en los datos.




### **1.1. Componentes principales de una Base de Datos**
* Datos: El contenido de la base de datos, que incluye información organizada en tablas. Por ejemplo, datos sobre usuarios, campañas electorales o resultados de encuestas.
* Tablas: Estructuras que organizan los datos en filas y columnas. Cada fila representa un registro (o entrada), y cada columna representa un atributo (o campo) del registro. Por ejemplo, una tabla de "Usuarios" podría tener columnas como "Nombre", "Email" y "Edad".
* Registros: Las filas de una tabla que contienen datos específicos. Cada registro representa un conjunto de información relacionado.
* Campos: Las columnas de una tabla, que definen los tipos de datos que se almacenan. Cada campo tiene un nombre y un tipo de datos (texto, número, fecha, etc.).
* Claves Primarias: Un campo que identifica de manera única (ID) cada registro en una tabla. Por ejemplo, wikidata o glottocode.
* Índices: Estructuras que mejoran la velocidad de las consultas al permitir un acceso más rápido a los datos. Son especialmente útiles en bases de datos grandes.

# **2. ¿Excel o CSV (Comma-Separated Values)?**

## **Semejanzas**
1. Almacenamiento de datos tabulados: Ambos formatos (CSV y Excel) se utilizan para organizar datos en filas y columnas, facilitando la visualización y análisis de la información.

2. Compatibilidad: Ambos formatos se pueden importar y exportar desde diversas aplicaciones, como bases de datos, herramientas de análisis y programas estadísticos, lo que los hace muy versátiles.

3. Uso común en Análisis de Datos: Estos formatos son muy utilizados en el análisis de datos y son compatibles con muchos lenguajes de programación y herramientas (como Python, R y pandas), lo que facilita su manipulación y estudio.

--------------------------

## **Diferencias**
1. **Formato de archivo**:
- Excel: Es un formato binario o basado en XML (XLS o XLSX) que puede contener múltiples hojas, gráficos, fórmulas, macros, y estilos de formato.
- CSV: Es un formato de texto plano donde los valores están separados por comas. No puede contener múltiples hojas ni formato avanzado.

2. **Capacidades y funcionalidades**:
- Excel: Soporta fórmulas, gráficos, tablas dinámicas, y una variedad de funcionalidades avanzadas de análisis de datos.
- CSV: Solo almacena datos sin ningún tipo de formato o funcionalidad adicional. No soporta fórmulas, gráficos ni otros elementos avanzados.

3. **Tamaño del archivo**:
- Excel: Los archivos pueden ser más grandes debido al almacenamiento de formato, fórmulas, y otros elementos adicionales.
- CSV: Generalmente, los archivos CSV son más pequeños y más ligeros, ya que solo contienen datos sin formato.

4. **Compatibilidad con software**:
- Excel: Requiere un software específico (Microsoft Excel) para abrir, visualizar y editar los archivos con todas sus funcionalidades.
- CSV: Puede ser abierto y editado con cualquier editor de texto (como Notepad) o software que soporte archivos de texto plano. Es más universal y fácil de compartir.

5. **Facilidad de lectura y escritura por programas**:
- Excel: Requiere librerías específicas para su manipulación en programación (por ejemplo, openpyxl o xlrd).
- CSV: Es muy fácil de leer y escribir en programación usando herramientas estándar en casi todos los lenguajes de programación (por ejemplo, csv en Python).

# **3.2.Gráfico de pastel (Pie charts)**

## Pueden revisar más en el siguiente enlace:
https://matplotlib.org/stable/plot_types/stats/pie.html#sphx-glr-plot-types-stats-pie-py

```python
# Importamos la librería Matplotlib
# import matplotlib.pyplot as plt

# Filtramos los partidos donde "tottenham" es el equipo local o visitante
partidos_tottenham = df_excel[(df_excel['HomeTeam'] == "Tottenham")]

# Creamos las variables para contar los partidos ganados, perdidos y empatados
ganados = 0
perdidos = 0
empatados = 0

# Iteramos sobre los partidos del Tottenham para contar los resultados y almacenarlos en las variables
# iterrows() es un método que permite iterar sobre las filas de un DataFrame
# index es el índice de la fila y row es la fila en sí
for index, row in partidos_tottenham.iterrows():
    if row['HomeTeam'] == "Tottenham": # Si el Tottenham es el equipo local en el partido
        if row['FTR'] == 'H': # Si el Tottenham ganó el partido
            ganados = ganados + 1 # Incrementamos el contador de partidos ganados
        elif row['FTR'] == 'A': # Si el Tottenham perdió el partido
            perdidos = perdidos + 1 # Incrementamos el contador de partidos perdidos
        else: # Si el partido terminó en empate
            empatados = empatados + 1 # Incrementamos el contador de partidos empatados

# Creamos las etiquetas, tamaños y colores para el gráfico de pastel
labels = ['Ganados', 'Perdidos', 'Empatados'] # Etiquetas para cada sección del gráfico
sizes = [ganados, perdidos, empatados] # Tamaños de cada sección del gráfico
colors = ['#88dc33', '#edbe35', '#FFC300'] # Colores para cada sección del gráfico
explode = (0.02, 0.05, 0.02) # Cuánto separar cada sección del gráfico

# Creamos el gráfico de pastel
plt.figure(figsize=(8, 8)) # Tamaño del gráfico
# plt.pie() crea un gráfico de pastel
# autopct='%1.1f%%' muestra el porcentaje en cada sección del gráfico
# shadow=True agrega sombra al gráfico
# startangle=140 rota el gráfico 140 grados en sentido antihorario
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=100)

# Añadir un título al gráfico
plt.title('Resultados de los Partidos como local del Tottenham (2023-2024)')
plt.axis('equal')  # Hace que el gráfico de pastel sea un círculo perfecto

# Guardamos el gráfico
plt.savefig('pastel_Tottenham.png', dpi=300)

# Mostramos el gráfico
plt.show()
```
