## **1. Diccionarios**
La programación se trata principalmente de resolver problemas del mundo real de la manera más eficiente posible.

En ese sentido, un diccionario en Python proporciona una forma más práctica de organizar los datos, es decir, funciona como una especie de "memoria asociativa" o almacenamiento de tipo **"clave:valor" (keys:values)**.

Es como tener un sistema de almacenamiento que te permite acceder rápidamente a datos utilizando **identificadores únicos (claves)** para recuperar información relacionada **(valores)**. Por lo tanto, el diccionario te permite describir dos piezas de datos y la relación que poseen.

```python
# Creamos un diccionario Notas donde las claves son los nombres de los alumnos y los valores son la nota que obtuvieron cada uno de ellos
Notas = {'Frank': 8,'Ronaldo': 18,'Serena': 15}

# Visualizamos los datos del diccionario Notas
print(Notas)
```

Podemos darnos cuenta de ciertas características sobre los diccionarios:

* Un diccionario está rodeado por llaves **{}** y los pares **"clave:valor"** están separados por comas.
* Un diccionario consta de uno o más pares "clave:valor". **La clave es el "identificador" o "nombre" que se utiliza para describir el valor.** Por ejemplo, en el diccionario 'Notas', uno de los pares "clave:valor" es 'Frank' : 8
- Las claves en un diccionario son únicas.
- La sintaxis para un par clave/valor es "CLAVE: VALOR".
- Las claves (por ejemplo, 'Frank') en un diccionario deben ser inmutables.
- Los valores (por ejemplo, 8) en un diccionario pueden ser cualquier objeto de Python.
- Un diccionario puede estar vacío.

Veamos qué tipos de datos funcionan y/o sirven como "claves" en un diccionaro. Por ejemplo, tenemos a los strings, los cuales sí funcionan como claves.

```python
# Los valores pueden ser listas en un diccionario

Notas_finales = {'Frank': [8, 10, 15], 'Susan': [7, 5, 12], 'Guido': [20, 18, 19]}

# Visualizamos los datos del diccionario Notas_finales

print(Notas_finales)
```

```python
# ¿Cómo obtener solo los keys de un diccionario?
# función .keys()
print(Notas_finales.keys())
```

```python
# ¿ Cómo obtener solo los values de cada key de un diccionario?
# función .values()
print(Notas_finales.values())
```

No obstante, en el caso de las listas, estas no funcionan como claves en un diccionario.

```python
diccionario_incorrecto = {['a', 'list']: 8}
```

### **1.2. ¿Cómo agregar elementos a un diccionario?**
Hay una manera muy simple de agregar un par **"clave:valor"** a un diccionario. Por favor, observa el siguiente fragmento de código:

```python
# La forma es nombre_diccionario["clave"] = valor
# Recuerda la clave o key siempre es un string
# Asimismo, el valor o vlaue puede ser todo tipo de datos y listas

a_dict['nombre'] = 'Juan'
a_dict['edad'] = 25
a_dict['ciudad'] = 'Bogotá'

print(a_dict)
```

Ten en cuenta que las claves del diccionario deben ser identificadores únicos para los valores en el diccionario. En ese sentido, **los pares "clave:valor" se sobrescriben si asignas un valor diferente a una clave existente**.

```python
# Solo el último valor asociado al key "Frank" se conserva
# Por lo tanto, los valores de un key pueden cambiarse

b_dict['Frank'] = 8
print(b_dict)
b_dict['Frank'] = 7
print(b_dict)
b_dict['Frank'] = 9
print(b_dict)
```

```python
print(b_dict) # Conserva el último valor asignado a la clave "Frank"
```

### **1.3. Contar elementos con un diccionario**
Los diccionarios son muy útiles para derivar estadísticas. Por ejemplo, podemos determinar fácilmente la frecuencia de cada letra en una palabra.

```python
# Creamos un diccionario vacio
frecuencia_letras = dict()

# Creamos una variable que contiene un string
palabra = 'otorrinolaringología'

# Recorremos un bucle for donde cada letra se convierte en una clave
for letra in palabra: # recorremos cada letra de la palabra
    if letra in frecuencia_letras: # si la letra ya está en el diccionario de frecuencias de letras
        frecuencia_letras[letra] = frecuencia_letras[letra] + 1 # sumamos 1 a la frecuencia de la letra
    else: # si la letra no está en el diccionario de frecuencias de letras
        frecuencia_letras[letra] = 1 # iniciamos la frecuencia de la letra en 1
print(frecuencia_letras) # imprimimos el diccionario de frecuencias de letras
```

```python
# Creamos una lista de palabras
texto_presidentes = "Obama fue presidente de los Estados Unidos. Castillo fue presidente del Perú. Piñera fue presidente de Chile. Peña Nieto fue presidente de México."
# Convertimos el texto a minúsculas
texto_presidentes_min = texto_presidentes.lower()
# Eliminamos los puntos
texto_presidente_depurado = texto_presidentes_min.replace(".", "")
# Convertimos el texto a una lista
lista_presidentes = texto_presidente_depurado.split()

# Creamos un diccionario vacío
frecuencia_palabra = {}

## Recorremos un bucle for donde cada palabra se convierte en una clave
for palabra in lista_presidentes:
    if palabra in frecuencia_palabra: # agregar 1 al diccionario si la clave ya existe
        frecuencia_palabra[palabra] = frecuencia_palabra[palabra] + 1
    else:
        frecuencia_palabra[palabra] = 1 # iniciamos la frecuencia de la palabra en 1

print(frecuencia_palabra)
```

## **2.Visualización con WordCloud**

Es importante mencionar que las nubes de palabras son útiles para visualizar palabras comunes en un texto o conjunto de datos, aunque por lo general solo son útiles como una descripción general.

Las nubes de palabras pueden ser particularmente útiles cuando se desea:

- Identificar rápidamente los temas o tópicos más importantes en un corpus de texto.
- Explorar patrones o tendencias en datos que contienen información textual.
- Comunicar las ideas o conceptos clave de una manera visualmente atractiva.

Sin embargo, es importante tener en cuenta que las nubes de palabras no proporcionan una comprensión más profunda de las palabras y frases que se utilizan. Por lo tanto, deben usarse junto con otros métodos para analizar e interpretar datos textuales.

---

- Importar la librería WordCloud

Esta librería nos permite generar nubes de palabras en Python

`from wordcloud import WordCloud`

- Importar la librería matplotlib

Esta librería nos permite generar gráficos en Python

`import matplotlib.pyplot as plt`

- Generar la nube de palabras con la librería WordCloud
  - width: ancho de la nube de palabras
  - height: alto de la nube de palabras
  - background_color: color de fondo de la nube de palabras
  - generate_from_frequencies: función que genera la nube de palabras a partir de las frecuencias de las palabras
  - frecuencia_palabra: diccionario con las frecuencias de las palabras
  - wordcloud: nube de palabras generada

- Personalizar la nube de palabras color
  - colormap: paleta de colores de la nube de palabras (los colores de colormap se pueden encontrar en https://matplotlib.org/stable/tutorials/colors/colormaps.html)

- Mostrar la nube de palabras

`plt.figure(figsize=(5,5))`: tamaño de la figura de la nube de palabras (ancho, alto)

`plt.imshow(wordcloud, interpolation="bilinear")`: mostrar la nube de palabras en la figura con interpolación bilineal (suaviza los bordes)

`plt.axis("off")`: no mostrar los ejes en la figura de la nube de palabras

`plt.show()`: mostrar la nube de palabras en la figura

- Guardar la nube de palabras en un archivo
`wordcloud.to_file("nube_palabras_presidentes.png")`
  - to_file: función que guarda la nube de palabras en un archivo
  - "nube_palabras_presidentes.png": nombre del archivo donde se guardará la nube de palabras

```python
# Importamos las librerías necesarias para realizar un gráfico de nubes de palabras
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generamos la nube de palabras
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(frecuencia_palabra)

# Mostramos la nube de palabras
plt.figure(figsize=(4,4))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

# Guardamos la nube de palabras en un archivo
wordcloud.to_file("nube_palabras_presidentes.png")
```

## **3. Visualización con Matplotlib (I)**

Matplotlib es una librería de Python ampliamente utilizada para la creación de gráficos y visualizaciones en dos dimensiones. Su flexibilidad y versatilidad permiten representar datos de manera clara y eficiente, siendo especialmente útil en proyectos de análisis de datos.

Matplotlib se destaca por su capacidad de personalización, lo que permite ajustar cada aspecto de un gráfico, desde colores, etiquetas y leyendas, hasta estilos de líneas y ejes. Es compatible con otras librerías de Python, como NumPy, WordCloud y pandas, lo que facilita la visualización de grandes conjuntos de datos.

Tipos de gráficos comunes con Matplotlib:
- Barras verticales y horizontales: Útiles para comparar cantidades entre categorías.
- Histogramas: Ideales para mostrar la distribución de una variable continua.
- Gráficos de pastel (pie chart): Perfectos para representar proporciones dentro de un conjunto.
- Gráficos de dispersión (scatter plot): Utilizados para explorar relaciones entre dos variables numéricas.
- Gráficos de líneas: Frecuentemente usados para visualizar tendencias en series temporales.
- Mapas: Con la extensión Basemap o combinando con otras librerías, es posible generar visualizaciones geográficas.

```python
# Visualizamos el texto
print(texto)
```

```python
# Convertimos el texto en minúscula

texto_minuscula = texto.lower()
print(texto_minuscula)
```

```python
# Borramos los signos de puntuación y los saltos de lineas \n

texto_depurado = texto_minuscula.replace(".", "").replace("\n", " ").replace(",", " ").replace("¿", " ").replace("?", " ").replace("¡", " ").replace("!", " ").replace(";", " ").replace("(", "").replace(")", "").replace("«", " ").replace("»", " ").replace(":", "")

###Visualizamos el texto
print(texto_depurado)
```

```python
# Creamos una lista de palabras
lista_palabras = texto_depurado.split()

# Visualizamos el texto
print(lista_palabras)
```

```python
# ¿Cuántos elementos tiene la lista?
cantidad_palabra = len(lista_palabras)

# Mostramos
print(cantidad_palabra)
```

```python
# ¿Cómo eliminamos elementos de una lista que no deseamos?
# función .remove()

lista_palabras.remove("1") # elimina el primer elemento "1"
lista_palabras.remove("4") # elimina el primer elemento "4"
lista_palabras.remove("4") # elimina el último elemento "4"

# Mostramos
print(lista_palabras)
```

```python
# Finalmente, ¿cuántos elementos tiene la lista?
cantidad_palabra = len(lista_palabras)

# Mostramos
print(cantidad_palabra)
```

```python
# Creamos un diccionario vacío
frecuencias = {}

# Recorremos un bucle for para contar el número de veces que aparece cada palabra de la lista
for palabra in lista_palabras:
    if palabra in frecuencias: ### El conteo de veces se agrega al diccionario vacio donde
        frecuencias[palabra] = frecuencias[palabra] + 1   ### cada palabra es un key y el número de veces es su valor
    else:
        frecuencias[palabra] = 1
```

```python
# visualizamos el diccionario
print(frecuencias)
```

```python
# Importamos las librerías necesarias para realizar un gráfico de nubes de palabras
# from wordcloud import WordCloud
# import matplotlib.pyplot as plt

# Generamos la nube de palabras
wordcloud = WordCloud(width=800, height=400, background_color="white", colormap='viridis').generate_from_frequencies(frecuencias)

# Mostramos la nube de palabras
plt.figure(figsize=(8, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

# Guardamos el gráfico
wordcloud.to_file("nube_palabras_hp.png")
```

```python
# Según el gráfico anterior, ¿cuáles son las palabras más frecuentes?
# preposiciones y los determinantes

# ¿Cómo eliminamos esos elementos?
# Ejemplo:
lista = [1, 2, 3, 4, 5]
lista_nueva = [2, 4]

for i in lista_nueva:
    lista.remove(i)
print(lista)
```

```python
# Creamos una lista de preposiciones
preposiciones = ['a', 'ante', 'bajo', 'cabe', 'con', 'contra', 'de', 'desde', 'durante', 'en', 'entre', 'hacia', 'hasta', 'mediante', 'para', 'por', 'según',
                  'sin', 'sobre', 'tras', 'versus', 'vía']
```

```python
# Creamos otra lista de determinantes
determinantes = ['el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'este', 'esta', 'estos', 'estas', 'ese', 'esa', 'esos', 'esas', 'aquel', 'aquella',
                  'aquellos', 'aquellas', 'mi', 'mis', 'tu', 'tus', 'su', 'sus', 'nuestro', 'nuestra', 'nuestros', 'nuestras', 'vuestro', 'vuestra',
                  'vuestros', 'vuestras', 'algún', 'ningún', 'cualquier', 'cada', 'otro', 'mismo', 'tanto', 'varios', 'poco', 'mucho', 'todo', 'demasiado']
```

```python
# Con el bucle for y la función .remove() eliminaremos los elementos de la lista preposiciones y determinantes de la lista palabras

# Crear una nueva lista filtrada
lista_filtrada = []

# Recorrer la lista de palabras
for palabra in lista_palabras:
    # Verificar si la palabra no está en preposiciones ni en determinantes
    if palabra not in preposiciones and palabra not in determinantes:
        # Agregar la palabra a la lista filtrada
        lista_filtrada.append(palabra)

# Mostrar el resultado
print(lista_filtrada)
```

```python
# Creamos un diccionario vacío
frecuencias_hp = {}

# Recorremos un bucle for para contar el número de veces que aparece cada palabra de la lista
for palabra in lista_filtrada:
    if palabra in frecuencias_hp: ### El conteo de veces se agrega al diccionario vacio donde
        frecuencias_hp[palabra] = frecuencias_hp[palabra] + 1   ### cada palabra es un key y el número de veces es su valor
    else:
        frecuencias_hp[palabra] = 1

print(frecuencias_hp)
```

```python
# Recuerden que el total de la lista es igual a la suma de los valores del diccionario

print(len(lista_filtrada)) # función len() mide el largo de la lista

print(sum(frecuencias_hp.values())) # función sum() realiza la suma de todos los valores del diccionario
```

```python
# ¿Cuántas palabras únicas presenta el texto de Harry Potter?

print(len(frecuencias_hp.keys()))
```

**`bar(x, y)`**: Dibuja un diagrama de barras verticales donde x es una lista con la posición de las barras en el eje X, e y es una lista con la altura de las barras en el eje Y.

- `clave_palabra`: contiene las palabras
- `veces_palabra`: contiene las frecuencias correspondientes

- Se crea un gráfico de barras con `plt.bar()`, donde las palabras están en el eje X y sus frecuencias en el eje Y.
- `plt.xticks(rotation=90)` rota las etiquetas de las palabras en el eje X para que se vean mejor.

- Se ajusta el límite del eje X con `plt.xlim(left=0)` para eliminar espacio en blanco.
- Finalmente, se guarda la imagen con `plt.savefig()` y se muestra con `plt.show()`.

```python
# Importamos las librerías que usaremos para graficar
# import matplotlib.pyplot as plt

# Extraer las palabras y sus frecuencias
clave_palabra = list(frecuencias_hp.keys())
veces_palabra = list(frecuencias_hp.values())

# Crear el gráfico de barras
plt.figure(figsize=(100, 10))  # Ajusta el tamaño del gráfico si es necesario
plt.bar(clave_palabra, veces_palabra, color='skyblue')  # Puedes cambiar el color

# Añadir etiquetas y títulos
plt.ylabel('Frecuencia')
plt.xlabel('Palabras')
plt.title('Frecuencia de palabras en Harry Potter')

# Rota las etiquetas del eje x para que se vean mejor
plt.xticks(rotation=90)

# Establecer los límites del eje x , esto permite que se elimine el espacio en blanco que se tenía antes del primer elemento del eje de x
plt.xlim(left=1)

# Guardamos
plt.savefig('ejemplo_1a.png', format='png')

plt.show()
```

### **3.2. Diagramas de barras horizontales**
`barh(x, y)`: Dibuja un diagrama de barras horizontales donde x es una lista con la posición de las barras en el eje Y, e y es una lista con la longitud de las barras en el eje X

```python
# Extraer las palabras y sus frecuencias

# Crear el gráfico de barras
plt.figure(figsize=(10, 35))  # Ajusta el tamaño del gráfico si es necesario
plt.barh(clave_palabra, veces_palabra, color='orange')  # Puedes cambiar el color

# Añadir etiquetas y títulos
plt.xlabel('Frecuencia')
plt.ylabel('Palabras')
plt.title('Frecuencia de palabras en la Lista')

# Establecer los límites del eje y
plt.ylim(bottom=0)

# Guardamos
plt.savefig('ejemplo_2.jpg', format='jpg')

plt.show()
```
