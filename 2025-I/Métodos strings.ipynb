### **`str.lower()`**

Este método convierte todo a minúsculas.

```python
palabra = "Aprobado"
texto = "El 22 de febrero se nos anunció que regresaríamos a Colombia."

print(palabra)
print(texto)
```

```python
print(palabra.lower())
print(texto.lower())
```

```python
print(palabra)
palabra_minuscula = palabra.lower()
print(palabra_minuscula)
```

### **`str.upper()`**

Este método convierte todo a mayúsculas.

```python
print(palabra.upper())
print(texto.upper())
```

### **`str.startswith()`**

Este método devuelve True si la cadena de caracteres empieza con el caracter o palabra indicada.

```python
print(palabra.startswith("A"))
print(texto.startswith("Ella"))
```

### **`str.endswith()`**

Este método devuelve True si la cadena de caracteres termina con el caracter o palabra indicada.

```python
print(palabra.endswith("A"))
print(texto.endswith("Colombia."))
```

### **`str.count()`**

Este método cuenta cuántas veces aparece un caracter o palabra en la cadena de caracteres principal.

```python
print(palabra.count("a"))
print(texto.count("re"))
```

### **`str.replace(viejo, nuevo)`**

Este método reemplaza un caracter o una palabra en la cadena de caracteres principal.

```python
print(palabra.replace("o","a"))
print(texto.replace("Colombia", "Perú"))
```

### **`str.split(sep)`**

Este método divide el string en partes, usando un separador o no.

```python
print(palabra.split())
print(texto.split())
```

----
### **`len(string)`**

Este función devuelve la cantidad de caracteres en el string, incluyendo espacios, signos de puntuación y saltos de línea.

```python
print(len(palabra))
print(len(texto))
```
